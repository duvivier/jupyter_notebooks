# Plots for Luke from Convect I test case
## (Currently without suppressing enhanced diffusivity)

```python
from matplotlib import pyplot as plt            # For plotting
import numpy as np
from netcdf_read import read_from_file          # Local module for reading data

# Place figures within document
%pylab inline
pylab.rcParams['figure.figsize'] = (12.0, 8.0) # Large figures
```

```python
resolutions = ['1m', '5m', '10m'];
timesteps = dict();
timesteps['1m']=['60s', '300s', '600s','1200s']
timesteps['5m']=['1200s']
timesteps['10m']=['60s', '300s', '600s','1200s']
start_day = dict();
end_day   = dict();
start_day['60s'] = 3*1496
end_day['60s'] = 3*1496
start_day['300s'] = 3*300
end_day['300s'] = 3*300
start_day['600s'] = 3*150
end_day['600s'] = 3*150
start_day['1200s'] = 3*75
end_day['1200s'] = 3*75

data  = dict();
for dz in resolutions:
    for dt in timesteps[dz]:
        ind = dz+'_'+dt
        data[ind] = read_from_file(start_day[dt], end_day[dt], 
                                   './test_output/'+ind+'.nc')        
```

```python
for dz in resolutions:
    ind = dz+'_1200s'
    plt.plot(data[ind]["w'T'"],data[ind]['zplot'],'-', linewidth=2.0, label=ind);
plt.title("w'T' averaged over "+data['10m_1200s']['day_for_plot'] + " (1200s time step)"); plt.xlabel("w'T'/surface value");
plt.ylabel("depth (m)"); plt.ylim((-65,0)); plt.legend(loc=4);
```

```python
dt = '1200s';
for dz in resolutions:
    ind = dz+'_'+dt
    plt.plot(data[ind]['time'], data[ind]['h'], '-', linewidth=2.0, label=ind)
plt.title("Boundary Layer Depth ("+dt+" time step)"); plt.xlabel("time (days)");
plt.ylabel("depth (m)"); plt.ylim((-100,0)); plt.legend();
```

```python
dz = '1m'
for dt in timesteps[dz]:
    ind = dz+'_'+dt
    plt.plot(data[ind]["w'T'"],data[ind]['zplot'],'-', linewidth=2.0, label=ind);

plt.title("w'T' averaged over "+data[dz+'_1200s']['day_for_plot'] + " (" + dz +" levels)"); plt.xlabel("w'T'/surface value");
plt.ylabel("depth (m)"); plt.ylim((-65,0)); plt.legend(loc=4);
```
